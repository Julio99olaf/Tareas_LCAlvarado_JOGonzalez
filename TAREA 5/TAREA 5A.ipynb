#### Problema 1

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st     # Librería estadística
%matplotlib inline
from scipy.optimize import fsolve
from scipy.stats import beta
from scipy.stats import triang
```

```python
# Límites de la función
a = 40
b = 49 
# Media, varianza 
mean = 44 
var = 9
# Valor para calcular probabilidad
d = 47


def alf_bet(ab):
    alf = ab[0]
    bet = ab[1]

    aa = ((a*bet+b*alf)/(alf+bet))-mean
    bb = ((alf*bet)*(b-a)**2)/(((alf+bet)**2)*(1+alf+bet))-var
    return np.array([aa,bb])

ab0= np.array([1,2])
ab = fsolve(alf_bet,ab0)

Alpha = ab[0]
Beta = ab[1]

print('Alpha =',Alpha)
print('Beta =',Beta)
```

```python
Prob = beta.cdf(d,Alpha,Beta,scale=(b-a),loc=a)
print('Probabilidad P(T<=d) es',Prob*100, '%')
```

```python
from matplotlib import pyplot as plt
x = np.linspace(30,60, 100)
betag= st.beta.pdf(x,Alpha,Beta,scale=(b-a),loc=a)
betag_acum= st.beta.cdf(x,Alpha,Beta,scale=(b-a),loc=a)

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.plot(x, betag,'r')
ax1.set_title('Densidad de probabilidad')
ax1.grid()
ax2.plot(x, betag_acum,'b')
plt.axvline(x=47, ymin=0, ymax=1,color='g')
ax2.set_title('Distribución acumulada')
ax2.grid()

plt.show()
```
